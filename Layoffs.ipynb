```python
import pandas as pd
import matplotlib.pyplot as plt
import altair as alt
```

```python
raw = pd.read_csv('Layoffs.csv')

dep = 'Department'
lay = 'Layoffs'
total = 'Total'
reduc = 'Percent of Full-time Faculty Cut'

raw[reduc] = raw[lay]/raw[total]

nozero = raw[raw[lay] > 0]

sortabs = nozero.sort_values(by=[lay], ascending = False)

noagg = sortabs[[dep, lay, reduc]][1:]

nozero = nozero[:-1] #take out total
```

```python
line = alt.Chart(noagg).mark_line(
    color="red",
    opacity=0.5
).encode(
    x=alt.X(dep+':O', sort=noagg[dep].tolist(), axis=alt.Axis(title="No. Layoffs")),
    y=reduc+':Q',
)

chart = alt.Chart(noagg).mark_bar(
    width=15,
    opacity=0.5
).encode(
    x=alt.X(dep+':O', sort=noagg[dep].tolist(), title="Percent of Fulltime Laid Off"),
    y=alt.Y(lay+':Q')
)

#alt.concat(chart1, line)
#(chart + line).properties(width=600)
alt.layer(chart, line).resolve_scale(y='independent')


#chart2 = alt.Chart(data).mark_point().encode(
#    x = 'date:T',
#    y = 'isotherm')
#
#chart3 = alt.Chart(data).mark_point(color='red').encode(
#    y = 'cfgi',
#    x='date:T')
#
#alt.concat(*(
#    alt.layer(chart2, chart3, title=f"WY = {val}").transform_filter(alt.datum.WY == val).resolve_scale(y='independent',x='independent')
#    for val in data.WY.unique()
#))

#(bar1 + bar2)
#source = data.iowa_electricity()
#
#alt.Chart(source).mark_bar(opacity=0.7).encode(
#    x='year:O',
#    y=alt.Y('net_generation:Q', stack=None),
#    color="source",
#)

#source = data.wheat()
#
#base = alt.Chart(source).encode(x='year:O')
#
#bar = base.mark_bar().encode(y='wheat:Q')
#
#line =  base.mark_line(color='red').encode(
#    y='wages:Q'
#)

#(bar + line).properties(width=600)

#base = alt.Chart(source).encode(
#    alt.X('month(date):T', axis=alt.Axis(title=None))
#)
#
#area = base.mark_area(opacity=0.3, color='#57A44C').encode(
#    alt.Y('average(temp_max)',
#          axis=alt.Axis(title='Avg. Temperature (Â°C)', titleColor='#57A44C')),
#    alt.Y2('average(temp_min)')
#)
#
#line = base.mark_line(stroke='#5276A7', interpolate='monotone').encode(
#    alt.Y('average(precipitation)',
#          axis=alt.Axis(title='Precipitation (inches)', titleColor='#5276A7'))
#)
#
#alt.layer(area, line).resolve_scale(
#    y = 'independent'
#)
```

```python

sortreduc = nozero.sort_values(by=[reduc], ascending=False)

line = alt.Chart(sortreduc).mark_line(
    color="red",
    size=3,
    #opacity=0.3
).properties(
    title="CCSF Potential Full-time Faculty Layoffs by Department"
).encode(
    x=alt.X(dep+':O', sort=sortreduc[dep].tolist(), axis=alt.Axis(title=None)),
    y=alt.Y(reduc+':Q', axis=alt.Axis(titleColor="red", format='%'))
)

chart = alt.Chart(sortreduc).mark_bar(
    width=15,
    opacity=0.9
).encode(
    x=alt.X(dep+':O', sort=sortreduc[dep].tolist()),
    y=alt.Y(lay+':Q', axis=alt.Axis(title="Number of Layoffs", titleColor="#4682B4", grid=True))
)

#alt.concat(chart1, line)
#(chart + line).properties(width=600)
alt.layer(line, chart).resolve_scale(y='independent').configure_view(
    strokeWidth=0
)
```
